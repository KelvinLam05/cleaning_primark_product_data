**Web Scraper**

Web Scraper is a chrome browser extension built for data extraction from web pages. Using this extension we can create a plan (sitemap) how a web site should be traversed and what should be extracted. Using these sitemaps the Web Scraper will navigate the site accordingly and extract all data. Scraped data later can be exported as CSV.

**Load packages and data**

```python
# Importing libraries
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv('/content/all_primark_products.csv')
```

```python
# Examine the data
df.head()
```

**Pre-processing**

```python
# Drop unwanted columns
df.drop(['web-scraper-order', 'web-scraper-start-url', 'pagination', 'product_url', 'product_url-href', 'image_url-src'], axis = 1, inplace = True)
```

We can use regular expressions to extract the subcategories.

```python
import re
```

```python
def manipulating_text(text_data):

  # Remove the word sub-categories
  text_data = re.sub('sub-categories', '', text_data)
  
  # Remove multiple characters
  text_data = re.sub(r'[]}[{":]', '', text_data)

  # Replace & with and 
  text_data = re.sub('&', 'and', text_data)

  return text_data
```

```python
df['subcategories'] = df['sub-categories'].apply(manipulating_text)
```

```python
df.head()
```

```python
# Drop the column sub-categories
df.drop(['sub-categories'], axis = 1, inplace = True)
```

```python
# Checking for missing values
df.isnull().sum().sort_values(ascending = False)
```

```python
# Drop missing values
df = df.dropna()
```

```python
# Reset the index
df = df.reset_index(drop = True)
```

```python
# Change the order of the Pandas DataFrame columns
df.insert(1, 'subcategories', df.pop('subcategories'))
```

**Exports the DataFrame to CSV format**

```python
df.to_csv('/content/all_primark_product_data_cleaned.csv', index = None, header = True) 
```
